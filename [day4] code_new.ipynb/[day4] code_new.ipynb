```python
from scipy import stats

count_normal = len(vib_normal)
min_normal = np.min(vib_normal)
max_normal = np.max(vib_normal)
mean_normal = np.mean(vib_normal)
med_normal = np.median(vib_normal)
mod_normal = stats.mode(vib_normal)[0][0]
var_normal = np.var(vib_normal)
std_normal = np.std(vib_normal)
range_normal = np.max(vib_normal) - np.min(vib_normal)
Q1_normal = np.quantile(vib_normal, 0.25)
Q3_normal = np.quantile(vib_normal, 0.75)
IQR_normal = np.quantile(vib_normal, 0.75) - np.quantile(vib_normal, 0.25)
LB_normal = np.quantile(vib_normal, 0.25) - 1.5 * (np.quantile(vib_normal, 0.75) - np.quantile(vib_normal, 0.25))
UB_normal = np.quantile(vib_normal, 0.75) + 1.5 * (np.quantile(vib_normal, 0.75) - np.quantile(vib_normal, 0.25))
RMS_normal = np.sqrt(np.mean(np.power(vib_normal, 2)))

print('count:', count_normal)
print('min:', min_normal)
print('max:', max_normal)
print('mean:', mean_normal)
print('median:', med_normal)
print('mode:', mod_normal)
print('variance:', var_normal)
print('standard deviation:', std_normal)
print('range:', range_normal)
print('Q1:', Q1_normal)
print('Q3:', Q3_normal)
print('IQR:', IQR_normal)
print('Lower Bound:', LB_normal)
print('Upper Bound:', UB_normal)
print('RMS:', RMS_normal)
```

```python
def manip(path, file):
    import numpy as np
    from scipy import stats
    import pandas as pd
    
    df = []
    for i in file:
        
        val = []
        with open(path+i, 'r', encoding = 'utf8') as f:
            rdr = csv.reader(f)
            for i in rdr:
                val.append(i[1])
        data = []
        for j in range(len(val)):
            if j < 9:
                data.append(val[j])
            else:
                data.append(float(val[j]))
                
        obj = {'dt': data[0],
               'count_v': len(data[9:]),
               'min_v': np.min(data[9:]),
               'max_v': np.max(data[9:]),
               'mean_v': np.mean(data[9:]),
               'meadian_v': np.median(data[9:]),
               'mode_v': stats.mode(data[9:])[0][0],
               'var_v': np.var(data[9:]),
               'std_v': np.std(data[9:]),
               'range_v': np.max(data[9:]) - np.min(data[9:]),
               'Q1': np.quantile(data[9:], 0.25),
               'Q3': np.quantile(data[9:], 0.75),
               'IQR': np.quantile(data[9:], 0.75) - np.quantile(data[9:], 0.25),
               'LB': np.quantile(data[9:], 0.25) - 1.5 * (np.quantile(data[9:], 0.75) - np.quantile(data[9:], 0.25)),
               'UB': np.quantile(data[9:], 0.75) + 1.5 * (np.quantile(data[9:], 0.75) - np.quantile(data[9:], 0.25)),
               'RMS': np.sqrt(np.mean(np.power(data[9:], 2)))
              }
        df.append(obj)
    dataDF = pd.DataFrame(df)
    return dataDF
```

```python
import os

path_normal = 'data/정상/'
files_normal = os.listdir(path_normal)
files_normal = [file for file in files_normal if file.endswith('.csv')]

df_normal = manip(path_normal, files_normal)
```
