```python
# 덧셈

x = numpy.array([1,2,3])
y = numpy.array([4,5,6])

a = x + y

print(a)
```

```python
# 뺼셈

x = numpy.array([1,2,3])
y = numpy.array([4,5,6])

a = x - y

print(a)
```

```python
# 나눗셈

x = numpy.array([1,2,3])
y = numpy.array([4,5,6])

a = x / y

print(a)
```

```python
# 곱셈

x = numpy.array([1,2,3])
y = numpy.array([4,5,6])

a = x * y

print(a)
```

```python
# 행렬곱

x = numpy.array([1,2,3])
y = numpy.array([4,5,6])

a = numpy.dot(x,y)

print(a)
```
